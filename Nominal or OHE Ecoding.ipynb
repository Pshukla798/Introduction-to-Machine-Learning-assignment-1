Nominal or OHE Ecoding

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df.head()
```

```python
# Create an instance of one hot encoder
ecd =OneHotEncoder()
```

```python
# fitting the ecd to the data frame and transform the categorical variable
ecoded=ecd.fit_transform(df[['sex','smoker','day','time']])
```

```python
ecd.fit_transform(df[['sex','smoker','day','time']]).toarray()
```

```python
import pandas as pd
eco_df=pd.DataFrame(ecoded.toarray(),columns=ecd.get_feature_names_out())
```

```python
eco_df
```

```python
pd.concat([df,eco_df])
```
