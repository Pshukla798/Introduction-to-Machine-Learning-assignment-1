```python
import numpy as np
x =np.array([1,2,3,4,5,6,7,8])
y= np.array([2,4,6,8,10,12,14,19])
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
# interpolate the data using the linear interpolation
x_new = np.linspace(1,5,10) ## create new x value
y_interp = np.interp (x_new,x,y) ## interpolate y values
print(y_interp)
```

```python
plt.scatter(x_new,y_interp)
```

```python
import numpy as np
x =np.array([1,2,3,4,5,6,7])
y= np.array([1,8,27,64,125,144,343])
```

```python
from scipy.interpolate import interp1d
```

```python
f=interp1d(x,y,kind='cubic')
```

```python
#interpolate the data
x_new == np.linspace(1,5,10)
Y_interp=f(x_new)
print(Y_interp)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x_new,Y_interp)
```

```python
# some sample data creation
x =np.array([1,2,3,4,5,6,])
y= np.array([1,4,9,16,25,36])
```

```python
from numpy.lib import polynomial
# interpoilate the dta using polynomial interpolation
p=np.polyfit(x,y,8) #fit a 2nd degree polynomial to the data
```

```python
x_new = np.linspace(1,5,10) ## creating x new values
y_interp = np.polyval(p,x_new) # interpolate y values
```

```python
plt .scatter(x,y)
```

```python
plt.scatter(x_new,y_interp)
```
